```python
"""
lets test if the normalization is good enough
"""
from normalization import fit_normalizer, transform_labels, save_normalizer
from astropy.table import Table
from astropy.io import fits
import galsim
import numpy as np
import os
```

```python
if os.path.exists("../../cosmos_catalog_train.fits"):
    with fits.open("../../cosmos_catalog_train.fits") as hdul:
        _cosmos_cat = hdul[1].data
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_1d(data, title="Data", bins=50):
    data = np.array(data)

    plt.figure()
    plt.hist(data, bins=bins)
    plt.title(title)
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.show()

raw_flux = _cosmos_cat['flux']
norm_flux = (raw_flux - raw_flux.mean()) / raw_flux.std()
inverse_norm_flux = (norm_flux * raw_flux.std()) + raw_flux.mean()
log_flux = np.log(raw_flux)
norm_log_flux = (log_flux - log_flux.mean()) / log_flux.std()
inverse_norm_log_flux = (norm_log_flux * log_flux.std()) + log_flux.mean()

plot_1d(raw_flux, "Raw Flux")
plot_1d(norm_flux, "Normalized Raw Flux")
plot_1d(inverse_norm_flux, "Inverse of Normalized Raw Flux")
plot_1d(raw_flux-inverse_norm_flux, "Diff of raw Flux and Inverse of Normalized Raw Flux")
plot_1d(log_flux, "Log Flux")
plot_1d(norm_log_flux, "Normalized Log Flux")
plot_1d(inverse_norm_log_flux, "Inverse of Normalized Log Flux")
plot_1d(log_flux-inverse_norm_log_flux, "Diff of raw Flux and Inverse of Normalized log Flux")
```
