```python
from astropy.io import fits
import numpy as np

with fits.open("/home/adfield/ShearNet/unit_tests/second_superbit_psf/benchmarking/psf_leakage/psf_leakage_results_sbpsf.fits") as hdul:
    data = hdul[1].data
    
print(f"s2n   — median: {np.nanmedian(data['s2n']):.2f}, mean: {np.nanmean(data['s2n']):.2f}")
print(f"T_gal — median: {np.nanmedian(data['T']):.4f}")
print(f"gal_flux_th — median: {np.nanmedian(data['gal_flux_th']):.2f}")  # will show 12258
print(f"flag==1 fraction: {np.mean(data['flag']==1):.3f}")
```
